```python
def fetch_processor(text):
    if text == 'Intel Core i7' or text == 'Intel Core i5' or text == 'Intel Core i3':
        return text
    else:
        if text.split()[0] == 'Intel':
            return 'Other Intel Processor'
        else:
            return 'AMD Processor'
```

```python
info_prices['Memory'] = info_prices['Memory'].astype(str).replace('\.0', '', regex=True)
info_prices["Memory"] = info_prices["Memory"].str.replace('GB', '')
info_prices["Memory"] = info_prices["Memory"].str.replace('TB', '000')
new = info_prices["Memory"].str.split("+", n = 1, expand = True)

info_prices["first"]= new[0]
info_prices["first"]=info_prices["first"].str.strip()

info_prices["second"]= new[1]

info_prices["Layer1HDD"] = info_prices["first"].apply(lambda x: 1 if "HDD" in x else 0)
info_prices["Layer1SSD"] = info_prices["first"].apply(lambda x: 1 if "SSD" in x else 0)
info_prices["Layer1Hybrid"] = info_prices["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
info_prices["Layer1Flash_Storage"] = info_prices["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

info_prices["first"] = info_prices['first'].str.replace(r'\D', '')

info_prices["second"].fillna("0", inplace = True)

info_prices["Layer2HDD"] = info_prices["second"].apply(lambda x: 1 if "HDD" in x else 0)
info_prices["Layer2SSD"] = info_prices["second"].apply(lambda x: 1 if "SSD" in x else 0)
info_prices["Layer2Hybrid"] = info_prices["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
info_prices["Layer2Flash_Storage"] = info_prices["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

info_prices["second"] = info_prices['second'].str.replace(r'\D', '')

info_prices["HDD"]=(info_prices["first"]*info_prices["Layer1HDD"]+info_prices["second"]*info_prices["Layer2HDD"])
info_prices["SSD"]=(info_prices["first"]*info_prices["Layer1SSD"]+info_prices["second"]*info_prices["Layer2SSD"])
info_prices["Hybrid"]=(info_prices["first"]*info_prices["Layer1Hybrid"]+info_prices["second"]*info_prices["Layer2Hybrid"])
info_prices["Flash_Storage"]=(info_prices["first"]*info_prices["Layer1Flash_Storage"]+info_prices["second"]*info_prices["Layer2Flash_Storage"])

info_prices.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
       'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
       'Layer2Flash_Storage'],inplace=True)
```

```python
X = info_prices.drop(columns=['Price'])
y = np.log(info_prices['Price'])
```

```python
X
```

```python
X.fillna(0, inplace=True)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=2)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor,ExtraTreesRegressor
from sklearn.svm import SVR
```

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = LinearRegression()

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```
