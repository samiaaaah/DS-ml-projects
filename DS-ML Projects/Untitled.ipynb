```python
#moviescsv
import numpy
import os
import pandas as pd
df=pd.read_csv('imdb-top-1000.csv')
df
```

```python
df_earn=df.groupby("Genre")["Gross"].sum()
df_sorted=df_earn.sort_values(ascending=False)
df_sorted.head(3)
```

```python
df_rate=df.groupby("Genre")["IMDB_Rating"].sum()
df_ratesort=df_rate.sort_values(ascending=False)
df_ratesort.head(1)
```

```python
df_=df[["Director","No_of_Votes"]]
df_sorted=df_.sort_values(by='No_of_Votes',ascending=False)
df_sorted.head(1)
```

```python
df_actor=df.groupby("Star1")["Series_Title"].sum()
df_actor
```

```python
df_rate=df.groupby("Genre")["IMDB_Rating"].mean()
df_ratesort=df_rate.sort_values(ascending=False).head(1)
df_ratesort
```
